```python
students = [
    {'이름': '김민수', '국어': 85, '영어': 90, '수학': 78},
    {'이름': '이영희', '국어': 92, '영어': 88, '수학': 95},
    {'이름': '박철수', '국어': 78, '영어': 85, '수학': 82}]

for s in students:
    s['총점'] = s['국어'] + s['영어'] + s['수학']
    s['평균'] = round(s['총점'] / 3, 2)
    print(f"총점: {s['총점']}점, 평균: {s['평균']}점")
```

```python
year = int(input("연도를 입력하세요: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year}년은 윤년입니다.")
else:
    print(f"{year}년은 평년입니다.")
```

```python
n = int(input("숫자를 입력하세요: "))

if n > 0:
    total = 0
    i = 1
    a = ""
    while i <= n:
        total += i
        if i == n:
            a += str(i)
        else:
            a += f"{i} + "
        i += 1
        
    print(f"{a} = {total}")
else:
    print("양수만 입력 가능합니다.")
```

```python
def print_gugudan(dan, start=1, end=9):
    if start > end:
        print("잘못된 범위입니다")
        return
    
    for i in range(start, end + 1):
        print(f"{dan} x {i} = {dan * i}")

print_gugudan(9,6,5)
```
