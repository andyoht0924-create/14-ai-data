```python
# 리스트는 어떻게 만들까?
# 형식

a = []
b = [1, 2, 3]
c = ["Life", "is", "too", "short"]
d = [1, 2, "Life", "is"]
e = [1, 2, ["Life", "is"]]

print(type(a))
print(len(a))
print(len(b))
print(len(c))
print(len(d))
print(len(e))
```

```python
# 리스트의 인덱싱
a = ["Life", "is", "too", "short"]
print(a[2])
print(a[-1])
print(a[0][-1])
```

```python
a = [1, 2, 3]
a[0] + a[2]
```

```python
a = [1, 2, 3, ["a", "b", "c"]]
print(a[-1])
print(a[-1][1])
```

```python
# 리스트의 슬라이싱
a = [1, 2, 3, 4, 5]
a[0:1+1]
```

```python
a[-2:]
```

```python
# 리스트 연산하기
# 1. 리스트 더하기
a = [1, 2, 3]
b = [4, 5, 6]
print(a + b)

# 2. 리스트 곱하기(반복)
a = [1, 2, 3]
print(a * 3)
```

```python
# 리스트의 수정과 삭제
# 1. 리스트의 값 수정하기
a = [1, 2, 3]
a[2] = 4
print(a)

# 2. 리스트의 요소 삭제하기
a = [1, 2, 3]
del a[1]
print(a)

a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```
