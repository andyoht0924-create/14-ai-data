```python
class Calculator:
    def __init__(self):
        self.result = 0
    
    def add(self, num):
        self.result += num
        return self.result
    
    def sub(self, num):
        self.result -= num
        return self.result
    
    def mul(self, num):
        self.result *= num
        return self.result
    
    def div(self, num):
        if num == 0:
            return "0으로 나눌 수 없습니다"
        self.result /= num
        return self.result

cal1 = Calculator()
print(cal1.add(10))
print(cal1.sub(2))
print(cal1.mul(5))
print(cal1.div(4))

cal2 = Calculator()
print(cal2.add(5))
print(cal2.mul(10))
print(cal2.sub(10))
print(cal2.div(5))
print(cal2.div(0))
```

```python
# 객체에 연산할 숫자 지정하기
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    
    def add(self):
        result = self.first + self.second
        return result

cal3 = FourCal()
print(type(cal3))
cal3.setdata(4, 2)

print(cal3.first)
print(cal3.second)
print(cal3.add())
```

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    
    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result

cal4 = FourCal()
cal4.setdata(4, 2)
print(cal4.add())
print(cal4.sub())
print(cal4.mul())
print(cal4.div())
```

```python
class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second
    
    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result

cal4 = FourCal(10, 20)
print(cal4.first)
print(cal4.second)
print(cal4.add())
print(cal4.sub())
print(cal4.mul())
print(cal4.div())
```

```python
# 클래스의 상속
class MoreFourCal(FourCal):
    pass

cal5 = MoreFourCal(4, 2)
print(cal5.add())
```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result

cal6 = MoreFourCal(10, 2)
print(cal6.add())
print(cal6.pow())
```

```python
# 메서드 오버라이딩
class MoreFourCal(FourCal):
    def add(self):
        result = self.first * self.second
        return result

cal4 = FourCal(5, 6)
print(cal4.add())
cal7 = MoreFourCal(5, 6)
print(cal7.add())
```
