```python
# 1. 일반적인 함수
# 입력값이 있고, 반환값이 있다.
def add(a, b):
    result = a + b
    return result

sum = add (3, 4)
print(sum)
```

```python
# 입렭값이 몇개가 될지 모를 때 어떻게 할까?
def add_many(*args):
    result = 0

    print(type(args))
    print(args)

    for i in args:
        result += i
    return result

result = add_many(1, 2, 3, 4, 5)
print(result)
```

```python
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result += i
    
    elif choice == "mul":
        result = 1
        for i in args:
            result *= i
    
    return result
```

```python
result = add_mul("add", 1, 2, 3, 4, 5)
print(result)
result = add_mul("mul", 1, 2, 3, 4, 5)
print(result)
```

```python
def create_profile(**kwargs):
    print("===== 프로필 정보 =====")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
# 함수의 반환값은 언제나 하나이다.
def add_and_mul(a, b):
    return a+b, a*b #하나 튜플(a+b, a*b) 괄호 생략
```

```python
result = add_and_mul(3, 4)
print(result)
```

```python
def add_and_mul(a, b):
    return a+b
    return a*b #하나만 가능

result = add_and_mul(3, 4)
print(result)
```

```python
def say_nick(nick):
    if nick =="학생":
        return
    print(f"너의 별명은 {nick}이다")

say_nick("선생")
say_nick("학생")
```

```python
# 매개변수의 초깃값 미리 설정하기
def say_myself(name, age, man=True):
    print(f"너의 이름은 {name}이다")
    print(f"너의 나이은 {age}이다")
    if man:
        print("남자입니다")
    else:
        print("여자입니다")

say_myself("홍길동", 25) #man=True로 설정해서 생략 가능
say_myself("김영희", 30, False)
```

```python
# 함수 안에서 선언한 변수의 효력 범위
a = 1
def vartest(a):
    a += 1 #함수 안에서만 가능

vartest(a)
print(a)
```

```python
a = 1
def vartest():
    global a
    a += 1

vartest()
print(a)
```

```python
# lambda 예약어
# 함수_이름 = lambda 매개변수1, 매개변수2, ... : 수행할 문장
def add(a, b):
    return a + b
# 같다
add = lambda a, b: a + b
```

```python
# 함수의 독스트링(docstring)
def add(a, b):
    """
    두 숫자를 더하는 함수
    
    parameters :
    a(int, float): 첫번째 숫자
    b(int, float): 두번째 숫자

    returns:
    int, float: 두 숫자의 합
    """
    return a + b

print(add.__doc__)
```
